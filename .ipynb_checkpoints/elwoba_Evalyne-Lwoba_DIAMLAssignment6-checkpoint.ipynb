```python
#data set
Titanic = pd.read_csv("titanic3.csv")

#function to fill age nan values with mean
def add_age(cols):
    age = cols[0]
    pclass = cols[1]
    if pd.isnull(age):
        return int(Titanic[Titanic["pclass"] == pclass]["age"].mean())
    else:
        return age
Titanic["age"] = Titanic[["age", "pclass"]].apply(add_age,axis=1) 

#change sex values from string to numeric
pd.get_dummies(Titanic["sex"])
sex_new = pd.get_dummies(Titanic["sex"],drop_first=True)
Titanic = pd.concat([Titanic,sex_new],axis=1)

#creating variables
x_survived = Titanic[["pclass","male","age"]]#predictor
y_survived = Titanic["survived"]#dependent

#decision tree plotting
model_tree = DecisionTreeClassifier()
model_tree.fit(x_survived,y_survived)
fig,ax = plt.subplots(figsize=(20,20))
tree.plot_tree(model_tree,fontsize=10,filled=True)
plt.show()

```

```python
#cross validation
model_tree = DecisionTreeClassifier()
cross_scores = cross_val_score(model_tree,x_survived,y_survived,cv=10)
print("The model accuracy before prunning is: ",round(cross_scores.mean()*100,2))
```

```python
# x_train,x_test,y_train,y_test = cross_validation.train_test_split(x,y,test_size=0.4,random_state=0)
count = []
score = []
for i in range(2,30):
    clf = tree.DecisionTreeClassifier(max_depth=i)
    # Perform 7-fold cross validation 
    scores = cross_val_score(clf,x_survived,y_survived, cv=7)
    count.append((i))
    score.append(round(scores.mean()*100,2))    
print(count)
print(score)

print("highest peformance accuracy score is",(max(score)))

#plotting scores and range
plt.gca().update(dict(title="DEPTH vs SCORE",xlabel="DEPTH",ylabel="SCORE"))
plt.rcParams["figure.figsize"]=(10,5)
plt.plot(count,score, color = "salmon",label="score against depth")
plt.scatter(5,max(score))
plt.legend()
plt.show()
```

```python
# Calculate the performance of the classifier versus the number of neighbors used and provide a 
# graphic to display the result. What is the optimal number of neighbors using cross-validation? (5 marks)
x_variables = Titanic[["pclass","male","age"]]
y_variables = Titanic["survived"]

x_train,x_test,y_train,y_test = train_test_split(x_survived,y_survived,test_size=0.2,random_state=1)

#in sample
knn = KNeighborsClassifier()
knn.fit(x_train,y_train)
score1 = cross_val_score(knn,x_train,y_train,cv=7)

#outsample
knn2 = KNeighborsClassifier()
knn2.fit(x_test,y_test)
score2 = cross_val_score(knn2,x_test,y_test,cv=7)

#accuracy
print("Default insample cross validation score is {:.3f}".format(score1.mean()*100))
print("Default outsample cross validation score is {:.3f}".format(score2.mean()*100))


#different neigh
k_range = range(1,31)
k_scores = []
k2_scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn,x_train,y_train,cv=3,scoring="accuracy")
    scores2 = cross_val_score(knn,x_test,y_test,cv=3,scoring="accuracy")
    k_scores.append(scores.mean())
    k2_scores.append(scores2.mean())

    
#plotting    
plt.plot(k_range,k_scores, label="in sample plot")
plt.plot(k_range,k2_scores,label = "out sample plot")
plt.scatter(2,max(k_scores),label = ("optimum score"))
plt.xlabel('Neighbour Number')
plt.ylabel("Accuracy of cross validation")
plt.title("KNeighbors vs Cross validation score")
plt.legend()
plt.show()

#print(k_scores)
print("optimal number of neighbours score is ",max(k_scores))
```

```python
# Evaluate the performance using different distance metrics. (5 marks)

k_metrics = ["chebyshev", "euclidean","hamming"]
k_scores = []
for i in k_metrics:
    knn = KNeighborsClassifier(n_neighbors=2,metric=i)
    scores = cross_val_score(knn,x_train,y_train,cv=5,scoring="accuracy")
    k_scores.append("{:.2f}".format(scores.mean()*100))
print("Distance metric scores are:",(k_scores))
```

```python
Wine_Red = pd.read_csv("winequality-red.csv", sep=";")
Wine_White = pd.read_csv("winequality-white.csv", sep=";")
```

```python
red_mean = pd.DataFrame(Wine_Red.mean(), columns=['mean_red'])
white_mean = pd.DataFrame(Wine_White.mean(), columns=['mean_white'])


cols = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol', 'quality']

#merging mean data frames
wine_merge = pd.DataFrame({"red_wine":red_mean.values.flatten(),"white_wine":white_mean.values.flatten()},index = cols)

wine_merge.plot.bar()
plt.gca().update(dict(title="RED AND WHITE WINE MEAN SCORES",ylabel="SCORE"))

plt.show()
```

```python
#correlation
#redwine
x_red = Wine_Red[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol','quality']]#independent
#whitewine
x_white = Wine_White[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol',"quality"]]#independent

corr_ = x_red.corr()
corre_ = x_white.corr()
print("Red wine correlation")
print(corr_[["quality"]])
print("\n")
print("White wine correlation")
print(corre_[["quality"]])
```

```python
# 4.3 Use Lasso and cross-validation to provide a plot of MSE against lambda and the parameter estimates
# versus lambda. How do the features selected by LASSO compare with an approach of setting a threshold on 
# the absolute correlation coefficient? (5 marks)

#redwine
x2_red = Wine_Red.drop(columns=['quality'],axis = 1)#independent
y_red = Wine_Red["quality"]#dependent

#whitewine
x2_white = Wine_White.drop(columns=['quality'],axis = 1)#independent
y_white = Wine_White["quality"]#dependent

#lasso redwine
lambda_red = np.logspace(-2,5)
red_mse = []
Rcoeff = []
Wcoeff = []
white_mse = []

for i in lambda_red:
    ModCl = linear_model.Lasso(alpha=i,random_state=3)
    Cl = ModCl.fit(x2_red,y_red)
    Cl2 = ModCl.fit(x2_white,y_white)
    #predicting
    red_pred = ModCl.predict(x2_red)
    white_pred = ModCl.predict(x2_white)
    #crossvalidate
    CrValidate = cross_validate(Cl,x2_red,y_red,cv=7,scoring="neg_mean_squared_error")
    CwValidate = cross_validate(Cl2,x2_white,y_white,cv=7,scoring="neg_mean_squared_error")
    #mean of mse
    Crscore = np.mean(CrValidate["test_score"])
    Crscore2 = np.mean(CwValidate["test_score"])
    #append redmse
    red_mse.append(abs(Crscore))
    coef_r = Cl.coef_
    Rcoeff.append(coef_r)
    #append white mse
    white_mse.append(abs(Crscore2))
    coef_w = Cl2.coef_
    Wcoeff.append(coef_w)

  
#plotting red wine
plt.plot(lambda_red,red_mse,label = "Red Wine MSE", color = "salmon",marker = "*")
plt.xscale("log")
plt.gca().invert_xaxis()
plt.gca().update(dict(title="PLOT OF MSE vs LAMBDA OF RED WINE",ylabel="MSE"))
plt.legend()
plt.grid()

```

```python
#plot white wine
plt.plot(lambda_red,white_mse,label = "WHITE WINE MSE", color = "blue",marker = "*")
plt.xscale("log")
plt.gca().invert_xaxis()
plt.gca().update(dict(title="PLOT OF MSE vs LAMBDA OF WHITE WINE",ylabel="MSE"))
plt.legend()
plt.grid()
```

```python
# parameter estimates
# red wine

RedC = pd.DataFrame(Rcoeff,columns=x2_red.columns,index = lambda_red)

#plotting 
RedC.plot()
plt.xscale = ("log")
plt.gca().invert_xaxis()
plt.legend(loc='lower right',prop={"size":10},ncol=4)
plt.gca().update(dict(title="PARAMETER ESTIMATES vs LAMBDA OF RED WINE",xlabel="Lambda",ylabel="Parameter Estimates"))
plt.grid()
```

```python
#white wine 
WhiteC = pd.DataFrame(Wcoeff,columns=x2_white.columns,index = lambda_red)

#plotting 
WhiteC.plot()
plt.xscale = ("log")
plt.gca().invert_xaxis()
plt.gca().update(dict(title="PARAMETER ESTIMATES vs LAMBDA OF WHITE WINE",xlabel="Lambda",ylabel="Parameter Estimates"))
plt.legend(loc='lower left',ncol=4)
plt.grid()
```
